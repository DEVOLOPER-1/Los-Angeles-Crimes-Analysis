# Exporting sub-csv files for instant display on my Dash Board

```python
from datetime import datetime

date_reported = df2["date_reported"].to_list()
print(date_reported[0])
date_occured = df2["date_occured"].to_list()
print(date_occured[1])

cases_numbers = []
differences = []
count = 0
for reported, occured in zip(date_reported, date_occured):
            if reported != "nan" and occured != "nan":

                try:
                    date_time_obj_r = datetime.strptime(reported, "%m/%d/%Y %I:%M:%S %p")

                    date_time_obj_o = datetime.strptime(occured, "%m/%d/%Y %I:%M:%S %p")

                    year_diff = abs(date_time_obj_r.year - date_time_obj_o.year)

                    month_diff = abs(date_time_obj_r.month - date_time_obj_o.month)

                    day_diff = abs(date_time_obj_r.day - date_time_obj_o.day)

                    total_diff_in_days = day_diff + (month_diff * 30) + (year_diff * 365)

                    differences.append(total_diff_in_days)
                    
                    count+=1
                    
                except:
                    count+=1
                    print(f"bad format {count}")


for index, row in df2.iterrows():
            case_num = row["division_of_records"]
            case_num = str(case_num)
            case_num = case_num[4:9]
            cases_numbers.append(case_num)

diff_date_bet_report_and_occurence_in_days_df = pd.DataFrame({'case_nuber': cases_numbers, 'differences': differences})
diff_date_bet_report_and_occurence_in_days_df.to_csv(r"C:\Users\youss\OneDrive\سطح المكتب\Kaggle\Los-Angeles-Crimes-Analysis\report_occurence.csv" , index=False)
```

```python
vict_ages = []
vict_sex_list = []
vict_descent_list = []
crime_code_identifier_list = []
for index , row in df2.iterrows():
    if str(row["vict_age"]).lower() != "nan"  and str(row["vict_sex"]).lower() != "nan" and str(row["vict_descent"]).lower() != "nan" and  str(row["crime_code_identifier"]).lower() != "nan" :
        if float(row["vict_age"]) > 0 or  float(row["vict_age"]) < 0:
            vict_ages.append(abs(row["vict_age"]))
            vict_sex_list.append(row["vict_sex"])
            vict_descent_list.append(row["vict_descent"])
            crime_code_identifier_list.append(row["crime_code_identifier"])

Age_Descent_Sex_crime_code = pd.DataFrame({'sex': vict_sex_list, 'age': vict_ages , 'victim Race': vict_descent_list , 'Crime against the victim' : crime_code_identifier_list})

Age_Descent_Sex_crime_code = Age_Descent_Sex_crime_code.clean_names()

print(Age_Descent_Sex_crime_code.columns)

types_convert_dict = {'sex': str, 'age': float , 'victim_race': str , 'crime_against_the_victim' : str}

Age_Descent_Sex_crime_code = Age_Descent_Sex_crime_code.astype(types_convert_dict)

Age_Descent_Sex_crime_code.to_csv(r"C:\Users\youss\OneDrive\سطح المكتب\Kaggle\Los-Angeles-Crimes-Analysis\vict_age_sex_descent_crime.csv" , index=False)
```

```python
from datetime import datetime
print(df2.columns)
#print(df2["occurence_time"])
time_list = ((df2["occurence_time"]).dropna(axis = 0 , how = 'any'  )).to_list()
str_time_list = []

for time in time_list:
    iterator = str(int(time))
#    print(iterator)
    if len(iterator) == 3 :
        iterator = "000" + iterator
    elif len(iterator) == 2 :
        iterator = "00" + iterator
    elif len(iterator) == 1:
        iterator = "000" + iterator
    str_time_list.append(iterator)
```

```python
hour_time_list = []
    
print(len(str_time_list))

for time in str_time_list:
        date_time_obj = datetime.strptime(time, "%H%M%S")
        formatted_time = date_time_obj.strftime("%H:%M")
        #print(formatted_time)
        hour_time_list.append(formatted_time)
print(len(hour_time_list))
precise_time_df = pd.DataFrame({'hr_min_time' : hour_time_list})
precise_time_df = precise_time_df.clean_names()
precise_time_df.to_csv(r"C:\Users\youss\OneDrive\سطح المكتب\Kaggle\Los-Angeles-Crimes-Analysis\precise_time_df.csv" , index=False)
```
